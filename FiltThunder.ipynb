```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
headers = ["Voltage", "DefocusU", "DefocusV", "DefocusTheta", "Cs", 
        "AmplitudeConstrast", "PhaseShift",
        "ImageName", "MicrographName", "CoordinateX", "CoordinateY", 
        "GroupNumber", "ClassNumber", 
        "q1", "q2", "q3", "q4", "StdRot1", "StdRot2", "StdRot3",
        "TransX", "TransY", "StdTransX", "StdTransY",
        "DefocusFactor", "StdDefocusFactor", "Score"]


def parse_thu(fname):
    thu = pd.read_table(fname, header=None, delim_whitespace=True)
    thu.columns = headers
    return thu
```

```python
r0 = parse_thu("Meta_Round_000.thu")
r31 = parse_thu("Meta_Round_031.thu")
```

```python
sns.regplot(r0["DefocusU"], r31["DefocusU"])
```

```python
sns.regplot(r0["DefocusTheta"], r31["DefocusTheta"])
```
